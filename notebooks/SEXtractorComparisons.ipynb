## Match Detected Objects

```python
def match(found,truth,width,height,scale):
    # Build a KD tree for the true source positions (in arcsecs) relative to the image center.
    num_truth = len(truth)
    xy_truth = np.empty((num_truth,2))
    xy_truth[:,0] = truth['dx']
    xy_truth[:,1] = truth['dy']
    kdtree = scipy.spatial.cKDTree(xy_truth)
    # Convert found object centroids from pixels relative to bottom-left corner to arcsecs
    # relative to image center.
    num_found = len(found)
    xy_found = np.empty((num_found,2))
    xy_found[:,0] = (found['X_IMAGE'] - 0.5*width - 0.5)*scale
    xy_found[:,1] = (found['Y_IMAGE'] - 0.5*height - 0.5)*scale
    # Find nearest true source to each found source.
    min_dist,truth_indices = kdtree.query(xy_found)
    # Returns tuple of:
    # - num_found bools indicating whether each object was matched with a truth object.
    # - num_matched truth-catalog indices for matched objects.
    # - num_matched Euclidean distances between matched objects in arcseconds.
    truth_matched = (truth_indices < num_truth)
    truth_indices = truth_indices[truth_matched]
    min_dist = min_dist[truth_matched]
    print 'Matched %d of %d detected objects, with median separation of %.2f pixels.' % (
        np.count_nonzero(truth_matched),num_found,np.median(min_dist)/scale)
    return truth_matched,truth_indices,min_dist
```

```python
ok,idx,dist = match(LSST_i_se,LSST_i,LSST_width,LSST_height,LSST_scale)
```
