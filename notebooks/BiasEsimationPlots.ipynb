### Read Simulation Outputs 

```python
output_catalog_path = '..'
def locate(name): return os.path.join(output_catalog_path,'%s.fits' % name)
```

```python
lsst_i = fits.open(locate('lsst_i'))
```

```python
#some interesting subsets of the simulation
cat = astropy.table.Table.read(lsst_i,hdu=1)
iso_gal = cat[cat['grp_size'] == 1] #isolated galaxies
grp_gal = cat[cat['grp_size'] != 1] #galaxies in a group of 2 or more. 

#'good' galaxies satisfy the reasonable criteron below 
good = cat[(cat['snr_iso'] > 10) & (cat['sigma_m'] > .2)]
iso_galg  = iso_gal[(iso_gal['snr_iso'] > 10) & (iso_gal['sigma_m'] > .2)] #'good' isolated galaxies
grp_galg  = grp_gal[(grp_gal['snr_grp'] > 10 )& (grp_gal['sigma_m'] > .2)] #'good' non isolated galaxies.

#gold sample galaxies 
gold = cat[(cat['ab_mag'] < 25.3)] 
iso_gold = iso_gal[(iso_gal['ab_mag'] < 25.3)] #isolated galaxies in the gold sample 
grp_gold = grp_gal[(grp_gal['ab_mag'] < 25.3)] #non isolated galaxies in the gold sample
```

```python
#number of galaxies in each category
print 'total number of galaxies is {0}'.format(len(cat))
print 'total number of isolated galaxies is {0}, they constitute {1:.3f}% of the total.'.format(len(iso_gal), float(len(iso_gal))/len(cat))
print 'total number of non-isolated galaxies is {0}, they constitute {1:.3f}% of the total.'.format(len(grp_gal), float(len(grp_gal))/len(cat))
print 'total number of good galaxies is {0}, they constitute {1:.3f}% of the total.'.format(len(good), float(len(good))/len(cat))
print 'total number of good isolated galaxies is {0}, they constitute {1:.3f}% of the total.'.format(len(iso_galg), float(len(iso_galg))/len(cat))
print 'total number of good non-isolated galaxies is {0}, they constitute {1:.3f}% of the total.'.format(len(grp_galg), float(len(grp_galg))/len(cat))
print 'total number of gold galaxies is {0}, they constitute {1:.3f}% of the total.'.format(len(gold), float(len(gold))/len(cat))
print 'total number of gold isolated galaxies is {0}, they constitute {1:.3f}% of the total.'.format(len(iso_gold), float(len(iso_gold))/len(cat))
print 'total number of gold non-isolated galaxies is {0}, they constitute {1:.3f}% of the total.'.format(len(grp_gold), float(len(grp_gold))/len(cat))
```

### Flux measurement biases 

```python
#similar plots but with bias on flux
def flux_biases_plots(cat,save=None):
    f_min = 0
    f_max = 3
    fig = plt.figure(figsize=(20,15))
    nrow,ncol = 1,3
    frac_bflux = cat['bias_f']/cat['flux']
    frac_bflux_grp = cat['bias_f_grp']/cat['flux']
    frac_bflux_g = good['bias_f']/good['flux']
    frac_bflux_grpg = good['bias_f_grp']/good['flux']
    frac_bflux_gold = gold['bias_f']/gold['flux']
    frac_bflux_grpgold = gold['bias_f_grp']/gold['flux']
    #
    plt.subplot(nrow, ncol, 1) 
    plt.hist(frac_bflux,range=[f_min,f_max],bins=100,histtype='step',label='iso bias')
    plt.hist(frac_bflux_grp,range=[f_min,f_max],bins=100,color='red',histtype='step',label='group bias')
    plt.legend()
    plt.yscale('log')
    plt.title('biases on flux for all galaxies')
    plt.xlabel('biases flux',fontsize=14)
    plt.ylabel('Number of galaxies',fontsize=14)
    #
    plt.subplot(nrow, ncol, 2) 
    plt.hist(frac_bflux_g,range=[f_min,f_max],bins=100,histtype='step',label='iso bias')
    plt.hist(frac_bflux_grpg,range=[f_min,f_max],bins=100,color='red',histtype='step',label='group bias')
    plt.legend()
    plt.yscale('log')
    plt.title('biases on $g_{1}$ for good galaxies')
    plt.xlabel('biases flux$',fontsize=14)
    plt.ylabel('Number of galaxies',fontsize=14)
    #
    plt.subplot(nrow, ncol, 3) 
    plt.hist(frac_bflux_gold,range=[f_min,f_max],bins=100,histtype='step',label='iso bias')
    plt.hist(frac_bflux_grpgold,range=[f_min,f_max],bins=100,color='red',histtype='step',label='group bias')
    plt.legend()
    plt.yscale('log')
    plt.title('biases on flux for gold galaxies')
    plt.xlabel('biases flux$',fontsize=14)
    plt.ylabel('Number of galaxies',fontsize=14)
```
